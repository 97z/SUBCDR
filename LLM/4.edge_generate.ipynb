```python
import json
from tqdm import tqdm
import re

mapped_file = "/root/tags_edge/new_elec_cell/elec_cell_style_tag_map.json"
with open(mapped_file, 'r', encoding='utf-8') as f:
    tag_mapping = json.load(f) 



def load_json(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        return json.load(f)

def fix_json_string(json_str):
    try:
       
        if json_str.startswith("Here is the output in JSON format:"):
            json_str = json_str[len("Here is the output in JSON format:"):].strip()
        if json_str.startswith("Here is the output in the required JSON format:"):
            json_str = json_str[len("Here is the output in the required JSON format:"):].strip()

        
        json_str = re.sub(r',\s*}', '}', json_str)
        return json_str
    except Exception as e:
        print(f"Error in fix_json_string: {e}")
        raise
def contains_chinese(text):
   
    return any('\u4e00' <= char <= '\u9fff' for char in text)
```

```python
def extract_and_map_tags(data):
    user_tags = {}

    for index, user in enumerate(data):
        user_id = user.get("user_id", index)  
        response_str = user.get("response", "").replace("\n", "").replace("    ", "")
        fixed_response = fix_json_string(response_str)

        try:
            
            if contains_chinese(fixed_response):
                raise ValueError("Response contains Chinese characters")

            
            if '"Style":' in fixed_response:
                emotion_content = fixed_response.split('"Style":')[1].split('},')[0]
                emotion_keys = json.loads(emotion_content + '}').keys()
                mapped_tags = {tag_mapping[tag] for tag in emotion_keys if tag in tag_mapping}

                if user_id is not None and mapped_tags:
                    user_tags[user_id] = mapped_tags

        except (json.JSONDecodeError, ValueError) as e:
            print(f"Error decoding index {index} JSON: {e}")

    return user_tags
```

```python

toy_file = "/root/tags_edge/new_elec_cell/elec_v11_summary.json"
video_file = "/root/tags_edge/new_elec_cell/cell_v11_summary.json"

toy_users_tags = extract_and_map_tags(load_json(toy_file))
video_users_tags = extract_and_map_tags(load_json(video_file))
```
