```python
def process_user(input_file, output_file, d1_user_num, d1_item_num, shared_num):
    with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
        for line in infile:
            data = line.strip().split(' ')
            user_id = int(data[0])
            user_src = int(data[1])
            #rating = data[2]
            #user_id, user_src, rating = line.strip().split(' ')
            if int(user_id) >= shared_num:
                user_id += d1_user_num + d1_item_num - shared_num

            outfile.write(f"{user_id}\t{user_src}\t{1}\n")
```

```python
import os

def process_user(input_file, d1_user_num, d1_item_num, shared_num, mode, final_output):
    #header = "tgt_user\tori_user\taspect\n"
    

    
    with open(final_output, 'a') as final_file, open(input_file, 'r') as infile:
        # if write_header:
        #     final_file.write(header)

        for line in infile:
            data = line.strip().split(' ')
            user_id = int(data[0])
            user_src = int(data[1])

            if user_id >= shared_num:
                user_id += d1_user_num + d1_item_num - shared_num

            final_file.write(f"{user_id}\t{user_src}\t{mode}\n")
          

    print(f"Processed {input_file} -> Merged into {final_output}.")
```
